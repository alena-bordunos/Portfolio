## 2. Предобработка данных

```python
data.isna().sum() # сколько пропущено значений в каждом столбце
```

```python
pd.DataFrame(round(data.isna().mean()*100,2)).style.background_gradient('coolwarm')
# выведем на экран количество пропусков в процентах
```

Пропуски в графе 'balcony' вероятнее всего означают отсутствие балкона в квартире, поэтому их можно заменить на 0. Пропуски в графе 'kitchen_area' (площадь кухни) также можно заменить на 0, так как этот показатель может отсутствовать, например, в квартирах-студиях. Пропуски в графе 'is_apartment' вероятнее всего отзначают отсутствие у объекта недвижимости статуса апартаментов, поэтому можно заменить их булевым значением False. Для удобства дальнейшей работы с данными заменим пропуски в графе 'locality_name' на значение 'неизвестно'. Пропусков в графе 'floors_total' совсем немного, данные этого столбца потребуются для анализа в дальнейшем, поэтому можно удалить строки с пропущенными значениями.

Пропуски в графе "высота потолков" нужно заполнить медианными значениями. Это будет гораздо информативнее для дальнейшего анализа, если перенести это действие на этап устранения аномалий.

```python
data['balcony'] = data['balcony'].fillna(0)
data['kitchen_area'] = data['kitchen_area'].fillna(0)
data['is_apartment'] = data['is_apartment'].fillna(False)
data['locality_name'] = data['locality_name'].fillna('неизвестно')
data = data.dropna(subset=['floors_total'])
```

Причины, которые могли привести к пропускам в данных: самая главная причина - человеческий фактор, продавцы не считают нужным заполнять некоторые необязательные параметры в объявлении. Этого можно избежать в дальнейшем, сделав эти параметры обязательными для размещения объявления, или, например, поднимать рейтинг объявления в поиске если пользователь заполнит информацию на 100%. Пропуски в графе days_exposition абсолютно нормальны и связаны с тем, что эти объявления еще активны и не сняты с продажи. Пропуски в столбцах, заполняемых автоматически могут быть связаны с тем, что парков или аэропорта просто нет на близком расстоянии до квартиры.

```python
data.isna().sum() # сколько пропущено значений в каждом столбце
```

```python
data['is_apartment'].unique() # посмотрим, можно ли присвоить апартаментам булев тип данных
```

```python
data['is_apartment'] = data['is_apartment'].astype('bool') # присвоим булев тип данных
data['balcony'] = data['balcony'].astype('int') # присвоим некоторым столбцам целочисленный тип данных
data['last_price'] = data['last_price'].astype('int')
data['floors_total'] = data['floors_total'].astype('int')
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
# изменим тип данных в столбце с датой для удобста дальнейшей работы с ним
```

Типы данных мы изменяем для удобства дальнейшей работы с ними, а также для удобства представления данных на диаграммах при проведении исследовательского анализа.

```python
data.info() # вызовем инфо, чтобы увидеть, что все сработало
```

```python
data.duplicated().sum() #проверка на явные дубликаты
```

```python
data['locality_name'].nunique()# проверим, сколько всего уникальных значений в графе locality_name 
```

```python
data['locality_name'].sort_values().unique() 
#список уникальных значений начеленных пунктов для поиска неочевидных дубликатов
```

```python
data['locality_name'] = data['locality_name'].str.lower()
```

```python
duplicated_name=(['поселок ', 'городского типа ', 'городской ', 
                  'коттеджный ', 'посёлок ', 'деревня ', 'садоводческое некоммерческое товарищество ', 
                  'садовое товарищество ', 'село ']
                )
correct_name = ''
for duplicate in duplicated_name:
    data['locality_name'] = data['locality_name'].str.replace(duplicate, correct_name)
    # убираем названия населенных пунктов для того, чтобы избавиться от дубликатов
```

```python
data['locality_name'].sort_values().unique() # проверяем результат работы цикла
```

```python
data['locality_name'].nunique() # еще раз выводим количество уникальных значений - стало на 56 меньше!
```

Теперь найдем и устраним редкие и выбивающиеся значения:

```python
print(data['ceiling_height'].describe()) # для начала посмотрим подробнее на столбец высота потолков
```

```python
data.boxplot(column='ceiling_height') # построим boxplot
plt.ylim(2,3)
```

```python
data['ceiling_height'].sort_values().unique() # посмотрим уникальные значения, чтобы определить природу выбросов
# тут явно видно, что свыше 20 метров это неверно введенная величина, которую можно разделить на 10
```

```python
data['ceiling_height'] = (
    data['ceiling_height'].where(data['ceiling_height']<20.0, other=data['ceiling_height']/10)
) #делим на 10 все потолки выше 20 метров
```

```python
ceiling_median = data['ceiling_height'].loc[data['ceiling_height'].isnull()==False].median()
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_median)
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height']>2.1, other=ceiling_median)
# на этом этапе можно заполнить пропуски в высоте потолков медианным значением
```

```python
data.isna().sum() # проверяем, исчезли ли пропуски
```

1. Заменены пропуски в тех местах, в которых это приемлемо для дальнейшего исследования
2. Изменены типы данных в ячейках для улучшения автоматизации анализа и наглядности предоставления данных

## 3. Добавление в таблицу новых столбцов

Добавим в таблицу новые столбцы со следующими параметрами:
1) цена одного квадратного метра;
2) день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
3) месяц публикации объявления;
4) год публикации объявления;
5) тип этажа квартиры (значения — «‎первый», «последний», «другой»);
6) расстояние до центра города в километрах (переведите из м в км и округлите до целых значений).

```python
data['last_price'] = data['last_price']/1000 # для удобного восприятия переведем в тысячи рублей
data['price_one_meter'] = data['last_price']/data['total_area'] 
data['weekday'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
# добавим столбцы в таблицу
```

```python
def floor_type(floor):
    if floor['floor'] ==1:
        return 'первый'
    elif floor['floor']==floor['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
data['floor_type'] = data.apply(floor_type, axis=1)
# создадим функцию для определения типа этажа и добавим столбец
```

```python
data['cityCenters_nearest_km'] = data['cityCenters_nearest']/1000
data['cityCenters_nearest_km'] = data['cityCenters_nearest_km'].round()
# добавим столбец расстояния до центра в километрах, применив округление после расчета
```

## 5. Расчет средней цены одного квадратного метра

Задание: посчитайте среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. Выделите населённые пункты с самой высокой и низкой стоимостью квадратного метра. Эти данные можно найти по имени в столбце locality_name.

```python

top_10_location = data['locality_name'].value_counts()[0:10]
# найдем 10 самых популярных населенных пунктов
top_10_location_price = (
    data.query('locality_name in @top_10_location.index')
    .pivot_table(index='locality_name', values='price_one_meter', aggfunc='median')
) # сделаем срез, используя предыдущую переменную, создадим сводную таблицу средних значений
print(top_10_location_price.sort_values('price_one_meter', ascending=False))
# выведем на экран сводную таблицу, отсортировав ее по убыванию
top_10_location_price.plot(kind='bar', grid=True) # для наглядности построим диаграмму

```

Самый дорогой квадратный метр - в городе Санкт-Петербург (104.6 тыс. руб.) и городе Пушкин (99.98 тыс. руб.), который является его частью, а также местом притяжения туристов. Самая низкая средняя цена - в городах Выборг (58.13 тыс. руб.) и Всеволожск (65.78 тыс. руб.), эти населенные пункты относятся к Ленинградской области, а также находятся на достаточной удаленности от центра Санкт-Петербурга, в сравнении с другими населенными пунктами, представленными в сводной таблице.

## 6. Расчет средней цены каждого километра

Задание: ранее вы посчитали расстояние до центра в километрах. Теперь выделите квартиры в Санкт-Петербурге с помощью столбца locality_name и вычислите среднюю цену каждого километра. Опишите, как стоимость объектов зависит от расстояния до центра города.

```python

data['center_km'] = data['cityCenters_nearest']/1000
data['center_km']=data['center_km'].round()
locality_spb = data.query('locality_name == "санкт-петербург"')

locality_spb_mean = locality_spb.pivot_table(index='center_km', values='price_one_meter', aggfunc='mean')
print(locality_spb_mean.sort_values('price_one_meter', ascending=False))
locality_spb_mean.plot(figsize=(10, 5), grid=True, style='o-', title = 'Заголовок');
```

На графике и сводной таблице видно, что средняя цена квадратного метра напрямую зависит от близости к центру. Самый дорогой квадратный метр (223 тысячи) - на нулевом удалении от центра, а самый дешевый в Санкт-Петербурге - 72 тысячи - на самом большом расстоянии (29й километр).
